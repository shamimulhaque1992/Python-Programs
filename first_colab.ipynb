```
data=[]
choice="random"
# When ever we call this function it will print an underline
def un_line():
  print("--------------------------------")

# When ever we call this function it will show us the menu for choosing the operations
def show_manu():
  un_line()
  print("Menu")
  un_line()
  print("1. Add Items")
  print("2. Mark as done")
  print("3. View list")
  print("4. Exit")
  un_line()

  #assigning the input in choice
  choice = input("Enter Your Choice: ")
  return choice

#until the input is 4 the loop will continue
while choice !="4":
  #assigning the return value of show_manu in choice
  choice=show_manu()
  if choice=="1":
    item=input("What is to be done? ")
    data.append(item)
    print("Item",item,"is added to the list!")
  elif choice=="2":
    item=input("What is to be marked as done? ")
    if item in data:
      data.remove(item)
      print("Inem",item,"is removed from the list")
    else:
      print("Item could not be found in the list")
  elif choice=="3":
     print("List of to do items")
     un_line()
     #until all the item is printed 
     for item in data:
       print(item) 
  elif choice=="4":
      print("GoodBuy!!!")
  else:
      print("Invalid Choice! Please chose one of 1, 2, 3 or 4")
   
```
